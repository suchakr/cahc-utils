```python
ans = pd.concat([ x[ x.apply( lambda r: bool(re.match(r'^\d+$', str(r[0]))) , axis=1) ].astype(
    {5:'int64'}).sort_values(5, ascending=False).assign(
    posn= lambda x: x[5].rank(ascending=False, method='min').astype('int64')
    , lead =lambda x : x[5].diff(-1).fillna(0).apply(int)
    , loss =lambda x : -x[5].diff(1).fillna(0).apply(int)

    )#.astype({'posn': 'int64', 'lead': 'int64'}) 
    for x in raw_pdfs ])
ans = ans.rename(columns={0:'sno', 1:'candidate', 2:'party', 3:'evm', 4:'postal', 5:'total', 6:'pct', 'cid':'cid', 'cname':'cname'})
ans = ans.astype({'sno': 'int64', 'candidate': 'string', 'party': 'string', 'evm': 'int64', 'postal': 'int64', 'total': 'int64', 'pct': 'float64', 'cid': 'int64', 'cname': 'string'})
ans = ans.drop(columns=['sno', 7, 8])
ans = ans.reset_index()
ans
```

```python
ans.groupby('party').agg(
    wins = ( 'posn' , lambda x: (x==1).sum()),
    # totalx = ('total', 'sum'),
    votes = ('total', 'sum'),
    pct = ('total', lambda x: (x.sum()/ans.total.sum())*100) ,
    runners = ( 'posn' , lambda x: (x==2).sum()),
    third = ( 'posn' , lambda x: (x==3).sum()),
    tail = ( 'posn' , lambda x: (x>3).sum()),
    contested = ('party', 'count'),
).sort_values(by=['wins', 'votes'], ascending=[False, False]).reset_index().style.background_gradient(
    cmap='Greens', subset=['wins', 'votes', 'pct' , 'runners']).format("{:.0f}%", subset=['pct']).format("{:,d}", subset=['wins', 'votes', 'contested', 'runners', 'third', 'tail'])
```

```python
ans[ans.posn==1].groupby('party').agg(
    wins = ( 'posn' , lambda x: (x==1).sum()),
    win_pct_mean = ( 'pct' , 'mean'),
    win_lead_mean = ( 'lead' , 'mean'),
).assign(
).sort_values(by=['wins', 'win_lead_mean', 'win_pct_mean'], ascending=[False, False, False]).reset_index().style.format("{:.0f}%", subset=['win_pct_mean']).format("{:.0f}", subset=['win_lead_mean'])
#  .background_gradient(
#      cmap='Greens', subset=['pct_mean', 'lead_mean']).format("{:.0f}%", subset=['pct_mean']).format("{:,d}", subset=['lead_mean']
# )
```

```python
winners = ans[ans.posn==1]
pvt = winners[['party', 'lead']].assign( 
    win_lead = lambda x: pd.cut(
      x.lead, 
      bins=[winners.lead.min(), 300, 500, 1000, 5000, 10_000, 20_000, 50_000, 100_000, winners.lead.max()], 
)).pivot_table(
    index='party', 
    columns='win_lead', aggfunc=len).fillna(0).astype('int64')

cols = pvt.columns
cols = [  f'{x.left:,d}-{x.right:,d}' for a ,x in cols]
pvt.columns = cols
pvt.plot.barh(stacked=True,width=.45, figsize=(8,5), title='Winning Margin', ylabel='Wins', xlabel='' )
pvt.style.background_gradient( cmap='Greens')
```
